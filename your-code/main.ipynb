## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data.head()
```

```python
data1 = data[data.apply(lambda x: (x['Country'] != 'United Kingdom') & (x['InvoiceDate'].month == 4) & (x['InvoiceDate'].year == 2011), axis=1)]
```

```python
data1 = data1.groupby('Country').sum().sort_values(by='Revenue', ascending = False).reset_index()
data1.head()
```

```python
import plotly.graph_objs as go
cf.go_offline()

trace1 = go.Bar(
    x=data1['Country'],
    y=data1['Quantity'],
    name='Total Quantity'
)
trace2 = go.Bar(
    x=data1['Country'],
    y=data1['Revenue'],
    name='Total Revenue'
)

data1f = [trace1, trace2]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data1f, layout=layout)
fig.iplot(filename='1')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
data3 = data[data['Description']=='PARTY BUNTING'].groupby('Country').mean().reset_index()
```

```python
data3.head()
```

```python
trace_comp0 = go.Scatter(
    x=data3.Quantity,
    y=data3.UnitPrice,
    mode='markers',
    marker=dict(size=12,
                line=dict(width=1),
                color='TURQUOISE'
               ),
    name='Country',
    text=data3.Country,
    )


data_comp = [trace_comp0]
layout_comp = go.Layout(
    title='Avg. Qty vs. Unit Price',
    hovermode='closest',
    xaxis=dict(
        title='Average Quantity',
        ticklen=5,
        zeroline=False,
        gridwidth=2,
    ),
    yaxis=dict(
        title='Average Unit Price',
        ticklen=5,
        gridwidth=2,
    ),
)
fig_comp = go.Figure(data=data_comp, layout=layout_comp)
fig_comp.iplot(filename='ap3')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
data4 = data[data.Country.isin(['EIRE','Germany','France', 'Netherlands'])]
```

```python
trace1 = go.Histogram(
    x=data4['Quantity'][data4.Country == 'EIRE'],
    name='EIRE'
)
trace2 = go.Histogram(
    x=data4['Quantity'][data4.Country == 'Germany'],
    name='Germany'
)
trace3 = go.Histogram(
    x=data4['Quantity'][data4.Country == 'France'],
    name='Fance'
)
trace4 = go.Histogram(
    x=data4['Quantity'][data4.Country == 'Netherlands'],
    name='Netherlands'
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(barmode='stack',
                  xaxis=dict(range=[0, 150])
                  )
fig = go.Figure(data=data, layout=layout)

fig.iplot(filename='overlaid histogram')
```
